```python
import string
def dropStudent():
    crn=input("Enter course CRN: ")
    sid=input("Enter Student's ID: ")#sid stands for student id
    crnList=[]# crnList is a list to maintain all the crn of courses
    studentId=[]#to store all the ids and crn as lists of registered students
    regStudents=[]#to store all the details of registered students
    courselist=[]#it is the list which consists all the courses information
    with open('coursesInfo.txt','r')as file:
        for line in file:
            temp=line.split(',')
            crnList.append(temp[0])
            courselist.append(temp)
    with open("registeredStudents.txt",'r') as file:
        for lines in file:
            temp=lines.split(',')
            studentId.append([temp[0],temp[1]])
            regStudents.append(temp)
    if crn in crnList and [crn,sid] in studentId:
        toDrop=input("Are you sure you want to drop this student from this course ? [Y/N] ")
        if(toDrop=='y' or toDrop=='Y'):
            print("Student %s,%s has been dropped from %s course successfully"%(sid,regStudents[studentId.index([crn,sid])][2],courselist[crnList.index(crn)][3]))
            regStudents.remove([crn,sid,regStudents[studentId.index([crn,sid])][2]])
            #now the regStudents were updated,need to update in file too
            with open('registeredStudents.txt','w')as file:
                for item in regStudents:
                    file.write(','.join([str(elem) for elem in item]))
    else:
        print('course does not exist or student is not registered in that course')
        
def registerStudent():
    crn=input("Enter course crn: ")
    courselist=[]
    #it is the list which consists all the courses information
    crnList=[]
    # crnList is a list to maintain all the crn of courses
    with open('coursesInfo.txt','r')as file:
        for line in file:
            temp=line.split(',')
            crnList.append(temp[0])
            courselist.append(temp)
    #now , the crnlist consists all the crn of courses present in coursesInfo.txt file
    #now , the courselist consists all the courses present in coursesInfo.txt file
    if crn in crnList:
        index=crnList.index(crn)
        #index is the index of crn in crnlist
        studentId=[]#to store all the ids and crn as lists of registered students
        with open("registeredStudents.txt",'r') as file:
            for lines in file:
                temp=lines.split(',')
                studentId.append([temp[0],temp[1]])
        #list of student ids is ready
        student=[]#to store student details
        student.append(crn)
        student.append(input("Enter student's ID: "))
        student.append(input("Enter student's name: "))
        #appending crn,student id ,student name to student list
        exist=False
        for item in studentId:
            if(item[0]==student[0] and item[1]==student[1]):
                print("student %s is already registered in this course"%(student[1]))
                exist=True
                
        if(not exist):
            #if the entered student is not registered for the entered course
            print("Student %s,%s has been registered in %s course successfully"%(student[1],student[2],courselist[index][3]))
            with open('registeredStudents.txt','a')as file:
                file.write(','.join([str(elem) for elem in student]))
            #appending student record into registeredStudents.txt file as a string
        
    else:
        print("course does not exist ")

def updateCourse():
    crn=input("Enter course crn: ")
    courselist=[]
    #it is the list which consists all the courses information
    crnList=[]
    # crnList is a list to maintain all the crn of courses
    with open('coursesInfo.txt','r')as file:
        for line in file:
            temp=line.split(',')
            crnList.append(temp[0])
            courselist.append(temp)
    #now , the crnlist consists all the crn of courses present in coursesInfo.txt file
    #now , the courselist consists all the courses present in coursesInfo.txt file
    if crn in crnList:
        index=crnList.index(crn)
        #index is the index of crn in crnlist
        print('1. Course Name')
        print("2. Instructor's name")
        print("3. Section Size")
        inp=int(input('What would you like to update ? '))
        if(inp==1):
            n=input("Enter course name: ")
            courselist[index][3]=n
        if(inp==2):
            courselist[index][4]=input("Enter Instructor's name: ")
        if(inp==3):
            courselist[index][5]=input("Enter Section size: ")
        print("course has been updated successfully ")
    else:
        print("course does not exist ")
            
    #the courseList was updated 
    #opening the coursesInfo.txt file in write mode will remove all its contents . so we can write freshly all the courses whichever required
    with open('coursesInfo.txt','w') as file:
        for lis in courselist:
            file.write(','.join([str(elem) for elem in lis]))
            #converting list to string joining by comma and writing it into file
            
    #now the txt file is updated


def removeCourse():
    crn=input("Enter course crn: ")
    courselist=[]
    #it is the list which consists all the courses information
    crnList=[]
    # crnList is a list to maintain all the crn of courses
    with open('coursesInfo.txt','r')as file:
        for line in file:
            temp=line.split(',')
            crnList.append(temp[0])
            courselist.append(temp)
    #now , the crnlist consists all the crn of courses present in coursesInfo.txt file
    #now , the courselist consists all the courses present in coursesInfo.txt file
    if crn in crnList:
        index=crnList.index(crn)
        #index is the index of crn in crnlist
        print("  CRN    Code  Sec No   Name                                     Instructor      Size Available") 
        for item in courselist[index]:
            print(item,end='   ')
        print("Are you sure you want to remove this course ? [Y/N] ")
        yesOrNo=input()
        if yesOrNo == 'y' or yesOrNo =='Y':
            print("Course has been successfully removed :) ")
            courselist.remove(courselist[index])
            #removing the course info from course list
    else:
        print("course does not exist ")
            
    #the courseList was updated by removing one course
    #opening the coursesInfo.txt file in write mode will remove all its contents . so we can write freshly all the courses whichever required
    with open('coursesInfo.txt','w') as file:
        for lis in courselist:
            file.write(','.join([str(elem) for elem in lis]))
            #converting list to string joining by comma and writing it into file
            
    #now the txt file is updated
        
def addCourse():
    #to add a new course into the coursesInfo.txt file
    newCourse='';
    #newCourse is an empty string and we are concatinating all the details of a course into it
    crnList=[]
    # crnList is a list to maintain all the crn of courses
    with open('coursesInfo.txt','r')as file:
        for line in file:
            temp=line.split(',')
            crnList.append(temp[0])
    #now , the crnlist consists all the crn of courses present in coursesInfo.txt file
            
    inp='';#is a string variable used to take and store the inputs
    while(len(inp)==0):
        inp=input('Enter course CRN: ')
        if inp in crnList:#checking if the entered crn is already present in the crnlist
            print("Serial number already exists")
            inp=''
    newCourse+=inp+','
    
    inp=''
    while(len(inp)==0):
        inp=input('Enter course code: ')
        if(len(inp)==0):
            print("course code can't be empty")
    newCourse+=inp+','
    
    inp=''
    while(len(inp)==0):
        inp=input('Enter Section Number: ')
        if(len(inp)==0):
            print("section number must be positive and non empty")
    newCourse+=inp+','
                  
    inp=''
    while(len(inp)==0):
        inp=input('Enter course name: ')
        if(len(inp)==0):
            print("course name can't be empty")
    newCourse+=inp+','
                  
    inp=''
    while(len(inp)==0):
        inp=input('Enter Instructor name: ')
        if(len(inp)==0):
            print("Instructure name can't be empty")
    newCourse+=inp+','

    inp=''
    while(len(inp)==0):
        inp=input('Enter Section size: ')
        if(len(inp)==0):
            print("section size must be positive and non empty")
    newCourse+=inp
    #now,we need to write this string into coursesInfo.txt file
    #for that i am opening it in append mode
    with open('coursesInfo.txt','a') as file:
        file.write(newCourse)
        file.write('\n')
    print("Course has been added successfully !!!")
    


def searchByNumber():
    #to search a course by course number,we need to input CRN
    crn=int(input("Enter course CRN: "))
    cnt=0
    lis=[]
    #'cnt' is a count variable to count no of matching records
    #and the 'lis' is a list of lists to store the matching records for further use
    with open('coursesInfo.txt','r') as file:
        #opened coursesInfo.txt in read mode for searching the record we needed
        for line in file:
            temp=line.split(',')
            # our txt file consists of string i.e., each row is a string
            #so we are splitting the string and stored in a list named 'temp' so that we can access each and every item using indexes
            if(int(temp[0])==crn):
                cnt+=1
                lis.append(temp)
                #if it is matched record ...appending it to "lis"
    if(cnt>0):
        # if the count is >0 means we found a matching record otherwise not
        print("Found a matching course")
        print(" CRN    Code  Sec No   Name                                       Instructor    Size Available") 
        #matched records were stored in lis.....so,printing the contents of lis 
        for lists in lis:
            for item in lists:
                print(item,end="   ")
    else:
        print("No matching course found")
        
def searchByCode():
    #to search a course by its code
    code=str(input("Enter course code: "))
    cnt=0
    lis=[]
    #'cnt' is a count variable to count no of matching records
    #and the 'lis' is a list of lists to store the matching records for further use
    with open('coursesInfo.txt','r') as file:
        #opened coursesInfo.txt in read mode for searching the record we needed
        for line in file:
            temp=line.split(',')
            # our txt file consists of string i.e., each row is a string
            #so we are splitting the string and stored in a list named 'temp' so that we can access each and every item using indexes
            #temp[1] consists code of the course
            if(str(temp[1])==code):
                cnt+=1
                lis.append(temp)
                #the matching records were appended to 'lis'
    if(cnt>0):
        # if the count is >0 means we found a matching record otherwise not
        
        print("    CRN    Code  Sec No   Name                                     Instructor      Size Available") 
        print('',end='   ')
        #matched records were stored in lis.....so,printing the contents of lis 
        for lists in lis:
            for item in lists:
                print(item,end="   ")
    else:
        print("course does not exist")
def searchByName():
    #to search a course by the name of the course
    name=input("Enter course name or part of it: ")
    cnt=0
    lis=[]
    #'cnt' is a count variable to count no of matching records
    #and the 'lis' is a list of lists to store the matching records for further use
    with open('coursesInfo.txt','r') as file:
        #opened coursesInfo.txt in read mode for searching the record we needed
        for line in file:
            temp=line.split(',')
            # our txt file consists of string i.e., each row is a string
            #so we are splitting the string and stored in a list named 'temp' so that we can access each and every item using indexes
            if name in temp[3].lower():
                #temp[3] consists course name
                cnt+=1
                lis.append(temp)
                #the matching records were appended to 'lis'
    if(cnt>0):
        # if the count is >0 means we found a matching record otherwise not
        
        print("  CRN    Code  Sec No   Name                                     Instructor      Size Available") 
        print('',end='   ')
        #matched records were stored in lis.....so,printing the contents of lis 
        for lists in lis:
            for item in lists:
                print(item,end="   ")
    else:
        print("course does not exist")
    
    
    
def searchCourse():
    print('1. By Course Number')
    print('2. By Course Code')
    print('3. By Course Name')
    ch=int(input("Enter your choice: "))
    if(ch==1):
        searchByNumber()
    elif(ch==2):
        searchByCode()
    else:
        searchByName()


def initializeCourseFile():
    #to create a new file named coursesinfo.txt
    #and to append some rows into it
    rows=["13068,ICS104,55,Introduction to Programming in Python and c,Saleh Alsaleh,30,0",
         "13080,ICS104,68,Introduction to Programming in Python and c,Saleh Alsaleh,30,0",
         "14292,ICS104,69,Introduction to Programming in Python and c,Saleh Alsaleh,30,0",
         "14293,ICS104,70,Introduction to Programming in Python and c,Saleh Alsaleh,30,0"]
   
    #rows is a list of strings which consists the data that need to write in newly created coursesinfo.txt
    with open('coursesInfo.txt','w') as file:
        #opened coursesInfo.txt file in write mode and the file object is named as file
        for row in rows:
            #for every string in rows we are writing into coursesInfo.txt file using file object "file"
            file.write(row)
            file.write('\n')
        file.close()
    #to create an empty registeredStudents.txt file
    with open("registeredStudents.txt",'w') as file:
        file.write('')
        file.close()

            
def printCourseInfo():
    #to print all the course information 
    #we have the courses information in coursesInfo.txt file
    #opening coursesInfo.txt file in read mode
    print("   CRN      Code    Sec No   Name                                          Instructor   Size Available") 
    print("",end="   ")
    with open('coursesInfo.txt','r') as file:
        for line in file:
            lis=line.split(',')
            for item in lis:
                print(item,end ="   ")
            
            

def main():
    initializeCourseFile()
    progress =True
    while(progress):
        print('University Register System')
        print('===================================')
        print('1.Print Courses info')
        print('2.Search for a course')
        print('3.Add a course')
        print('4.Remove a course')
        print('5.Update a course')
        print('6.Register a student in a course')
        print('7.Drop a student from a course')
        print('8.Exit')
        print('===================================')
        choice= int(input('Enter your choice: '))
        if(choice==1):
            printCourseInfo()
        elif(choice==2):
            searchCourse()
        elif(choice==3):
            addCourse()
        elif(choice==4):
            removeCourse()
        elif(choice==5):
            updateCourse()
        elif(choice==6):
            registerStudent()
        elif(choice==7):
            dropStudent()
        else:
            progress=False

            
if __name__=="__main__":
    main()
```
