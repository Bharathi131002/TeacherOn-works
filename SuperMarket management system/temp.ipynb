```python
import csv
import shutil

def initializeFile():
    rows=[[123,'pen',1.5,10,0.3],
     [124,'pencil',1,15,0.1],
     [125,'book',50,2,0.0]]
    with open('store.txt','w') as csvfile:
        csvwriter =csv.writer(csvfile)
        csvwriter.writerows(rows)
        csvfile.close()


def printInfo():
    cnt=0
    lis=[]
    with open('store.txt',mode='r')as file:
        csvFile=csv.reader(file)
        for lines in csvFile:
            if(len(lines)>0):
                cnt+=1
                lis.append(lines)
        print('Total %d products:'%(cnt))
        print('-----------------------------------')
        for record in lis:
            print('Id:        %s'%(record[0]))
            print('title:     %s'%(record[1]))
            print('price:     %s'%(record[2]))
            print('quantity:  %s'%(record[3]))
            print('discount:  %s'%(record[4]))
            print('-----------------------------------')



def addNewProd():
    lis=[]
    flis=[]
    added=False
    lis.clear()
    x=input('Enter Id: ')
    lis.append(x)
    x=input('Enter name: ')
    lis.append(x)
    x=input('Enter price: ')
    lis.append(x)
    x=input('Enter quantity: ')
    lis.append(x)
    x=input('Enter discount: ')
    lis.append(x)
    with open('store.txt',mode='r')as file:
        csvFile=csv.reader(file)
        for lines in csvFile:
            if(len(lines)>0 ):
                if(lines[0]==lis[0] and lines[1]==lis[1] ):
                    lines[3]=str(int(lines[3])+int(lis[3]))
                    added=True
                flis.append(lines)
    #writing updated contents to tempfile and rename it
    with open('tempstore.txt','w') as csvtemp:
        csvwriter=csv.writer(csvtemp)
        csvwriter.writerows(flis)
    shutil.move('tempstore.txt','store.txt')
    if(added==False):
        with open('store.txt','a')as csvfile:
            csvwriter=csv.writer(csvfile)
            csvwriter.writerow(lis)
            csvfile.close()
    print("New product was added successfully")


def deleteProd(id,nr):
    flis=[]
    cd=True
    found=False
    with open('store.txt',mode='r')as file:
        csvFile=csv.reader(file)
        for lines in csvFile:
            if(len(lines)>0 ):
                if(int(lines[0])==id and int(lines[3])>=nr ):
                    lines[3]=str(int(lines[3])-nr)
                    found=True
                    if(int(lines[3])>0):
                        flis.append(lines)
                        cd=False
                        continue;
                if(int(lines[3])>0):
                    flis.append(lines)
                
    if(found):
        if(cd):
            print('product with id %d was completely removed'%(id))
        else:
             print('product with id %d was reduced by %d quantity'%(id,nr))
            
            
    else:
        print("Id not found")
        
    with open('tempstore.txt','w') as csvtemp:
        csvwriter=csv.writer(csvtemp)
        csvwriter.writerows(flis)
    shutil.move('tempstore.txt','store.txt')
                    
    
def removeProd():
    id=int(input('Enter product id: '))
    nr=int(input('Enter quantities to remove: '))
    c=input('Are you sure you want to remove (y/n): ')
    print('-------------------------------------------------')
    if(c=='y' or c== 'Y'):
        deleteProd(id,nr)


    

def searchById():
    id=input('Enter the id: ')
    cnt=0
    lis=[]
    with open('store.txt',mode='r')as file:
        csvFile=csv.reader(file)
        for lines in csvFile:
            if(len(lines)>0 and lines[0]==id):
                cnt+=1
                lis.append(lines)
    print('Matched products (%d)'%(cnt))
    print('=======================================')
    for record in lis:
        print('Id:        %s'%(record[0]))
        print('title:     %s'%(record[1]))
        print('price:     %s'%(record[2]))
        print('quantity:  %s'%(record[3]))
        print('discount:  %s'%(record[4]))
        print('-----------------------------------')
            
def searchByName():
    name=input('Enter the name: ')
    cnt=0
    lis=[]
    with open('store.txt',mode='r')as file:
        csvFile=csv.reader(file)
        for lines in csvFile:
            if(len(lines)>0 and lines[1][:len(name)]==name):
                cnt+=1
                lis.append(lines)
    print('Matched products (%d)'%(cnt))
    print('=======================================')
    for record in lis:
        print('Id:        %s'%(record[0]))
        print('title:     %s'%(record[1]))
        print('price:     %s'%(record[2]))
        print('quantity:  %s'%(record[3]))
        print('discount:  %s'%(record[4]))
        print('-----------------------------------')

    
def searchProd():
    print('Choose an option')
    print('=============================')
    print('1. Search by id')
    print('2. Search by name')
    print('3. Return to Main Menu')
    print('=============================')
    c=int(input('Enter your choice: '))
    if(c==1):
        searchById()
    elif(c==2):
        searchByName()
#     else:
#         #return to main menu



    

def printProdByRange(l,h):
    cnt=0
    lis=[]
    with open('store.txt',mode='r')as file:
        csvFile=csv.reader(file)
        for lines in csvFile:
            if(len(lines)>0 and float(lines[2])>=l and float(lines[2])<=h):
                cnt+=1
                lis.append(lines)
        print('Matched products (%d):'%(cnt))
        print('==========================================')
        for record in lis:
            print('Id:        %s'%(record[0]))
            print('title:     %s'%(record[1]))
            print('price:     %s'%(record[2]))
            print('quantity:  %s'%(record[3]))
            print('discount:  %s'%(record[4]))
            print('-----------------------------------')
    
def filterByPriceRange():
    range=input("Enter a price range: ").split('-')
    printProdByRange(float(range[0]),float(range[1]))
    

    

def modifyProd():
    id=input('Enter id: ')
    flis=[]
    found=False
    
    with open('store.txt',mode='r')as file:
        csvFile=csv.reader(file)
        for lines in csvFile:
            if(len(lines)>0 ):
                if(lines[0]==id ):
                    print('==================================\n Found product: Enter its info')
                    lines[1]=input('Enter name: ')
                    lines[2]=input('Enter price: ')
                    lines[3]=input('Enter quantity: ')
                    lines[4]=input('Enter discount: ')
                    print("product was modified successfully")
                    found=True
                flis.append(lines)
    if(not found):
        print('Product not found')
    #writing updated contents to tempfile and rename it
    with open('tempstore.txt','w') as csvtemp:
        csvwriter=csv.writer(csvtemp)
        csvwriter.writerows(flis)
    shutil.move('tempstore.txt','store.txt')
    
def main():
    initializeFile()
    progress=True
    while(progress):
        print('SUPER MARKET SYSTEM')
        print('===============================')
        print('1. Print Products info')
        print('2. Search a product')
        print('3. Add a product')
        print('4. Remove a product')
        print('5. Modify a product')
        print('6. Filter by price')
        print('7. exit')
        print('===============================')
        ch = int(input('Enter your choice: '))
        if(ch==1):
            printInfo()
        elif(ch==2):
            searchProd()
        elif(ch==3):
            addNewProd()
        elif(ch==4):
            removeProd()
        elif(ch==5):
            modifyProd()
        elif(ch==6):
            filterByPriceRange()
        else:
            progress=False
    
    
    
if __name__=="__main__":
    main()
    

```
